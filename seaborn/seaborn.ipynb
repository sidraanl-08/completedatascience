```python
#scatter plot

import seaborn as sb 


sb.get_dataset_names() #seaborn has builtin datasets
penguins = sb.load_dataset( 'penguins')
print(penguins)

sb.scatterplot(data = penguins, x = 'flipper_length_mm' , y = 'body_mass_g', hue='island')
```

## Styling In SeaBorn ðŸŒŠ


- set_style(): It controls appearance and style of plots ticks, white, dark, whitegrid, darkgrid, despine(to remove spines)
- set_context(): It controls scaling of plot elements (fonts, lines, labels) to make plots suitable for notebooks, talk, paper, poster
- palette: Set1, Set2, Set3 are for changing the graph colors
- palette: dark, deep, muted, pastel, colorblind, bright to change tone of the colors
- palettetype: qualitative, sequential, diverging

```python
#demonstrating set_style()

sb.set_style('ticks')
sb.scatterplot(data = penguins, x = 'flipper_length_mm' , y = 'body_mass_g', hue='island')
sb.despine()


#demonstrating set_context()

sb.set_context('paper')
sb.scatterplot(data = penguins, x = 'flipper_length_mm' , y = 'body_mass_g', hue='island', palette='dark')



```

```python
#obtaining more readability with scatter plot

sb.scatterplot(data=penguins, x = 'species', y = 'body_mass_g', hue = 'island', style='sex', alpha=0.5)
```

# Understanding Other Types Of Plots In SeaBorn ðŸŒŠ

```python
#obtaining more readaility with stripplot and using dodge and jitter

sb.stripplot(data=penguins, x = 'species', y = 'body_mass_g', hue = 'island', dodge=True, jitter = True)
```

```python
#randomly distributing data with swarm plot

sb.set_context('talk')
sb.swarmplot(data = penguins,  x = 'species', y = 'body_mass_g', hue = 'island', palette='dark')
sb.despine()
```

```python
#relationship of two variables with regression plot
#it sets a linear line through data for getting more insights

sb.regplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', color='red')
```

```python
#demonstrating with lineplot

sb.set_style('whitegrid')
sb.lineplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', color='orange', hue = 'island', style='sex')
```

```python
#demonstrating join plot to compare two different plots like a scatter plot or a histogram

sb.jointplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', color='purple', hue='sex')

sb.jointplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', color='purple')

sb.jointplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', color='purple', hue='sex', kind = 'kde')
```

```python
#demonstrating barplot

sb.barplot(data = penguins, x = 'species', y = 'body_mass_g', color='purple', hue='sex', palette=['pink', 'blue'])
```

```python
#demonstrating boxplot

sb.boxplot(data = penguins, x = 'species', y = 'body_mass_g', color='orange', hue='sex')
```

```python
#demonstrating violin plot

sb.violinplot(data = penguins, x = 'species', y = 'body_mass_g', color='skyblue', hue='sex')

```

```python
sb.violinplot(data = penguins, x = 'species', y = 'body_mass_g', color='red', hue='sex', split=True)
```

```python
#demonstrating swarmplot and violinplot together

sb.violinplot(data = penguins, x = 'species', y = 'body_mass_g', palette='pastel')
sb.swarmplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'island', size = 3, palette='dark')
```

```python
#demonstrating KDEplot kernel density estimation
#same as a histogram but kdeplots are smoother than them
#used to estimate probability density of function of continuous and non-parametric data
#shows concentrated and dense data through peaks along its distribution

sb.kdeplot(data = penguins, x = 'body_mass_g', hue='species', fill=True)
```

```python
#demonstrating Heatmap
#its a 2d color coded matrix that shows relationship between two variables
#highlight missing data -> sb.heatmap(df.isnull())

penguins.head()
columns = ['bill_length_mm', 'bill_depth_mm', 'body_mass_g']
penguins[columns].corr()

sb.heatmap(data = penguins[columns].corr(), annot=True, linewidth= 3, linecolor='black') #annot will denote the heatmap

```

```python
#demonstrating rugplot

sb.rugplot(data = penguins, y = 'body_mass_g',  hue = 'species', palette = 'Set2', height = 0.2)
```

```python
#demonstrating pairplot
#parallel visualization of all numerical columns in data

sb.pairplot(data = penguins, hue = 'species', palette='Set2') #doesnt raise error if columns are not defined and gives pairplots of all data
```

```python

sb.pairplot(data = penguins[columns]) #now this will give pairplots of numerical columns only
```

```python
#demonstrating pairgrid
#a subplot grid for plotting pairwise relationships in a data 

graph = sb.PairGrid(data  = penguins, hue = 'sex', palette = 'Set2')
graph.map_upper(sb.scatterplot)
graph.map_lower(sb.kdeplot)
graph.map_diag(sb.histplot)
graph.add_legend()
```
