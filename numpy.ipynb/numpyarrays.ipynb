## 1. Creating NumPy Arrays

### With Python Objects (list, tuple):

```python
#using a python object, list to create an array object

import numpy as np

arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr)) #here the arr is numpy.ndarray object but type of np.array will be builtin function

listarr = np.array([[1,2,3], [5,8,5], [0,3,1]])
print(listarr)
print(listarr.dtype)
print(listarr.shape)
print(listarr.size)

#using a python object, tuple to create an array object 

arr2 = np.array((10,20,30,40,50))
print(arr2)

```

### With Intrinsic Objects: 

- We can use arange, ones, zeroes to make arrays with NumPy intrinsic objects

```python
#creating an array with np.zeroes()

zeroes = np.zeros((2,5))
print(zeroes)
print(zeroes.dtype)
print(zeroes.shape)
print(zeroes.size)

#creating an array with np.ones()

ones = np.ones((2,6))
print(ones)
print(ones.dtype)
print(ones.shape)
print(ones.size)


#creating an array with np.arange() 

rng = np.arange(15)
print(rng)
print(rng.dtype)
print(rng.shape) #it gives here 15, coz NumPy returns arrays as tuples, so its telling its a tuple array not number 15
print(rng.size)


#creating an array using np.linspace()

lsp = np.linspace(1,5,12) #the range of number will be 1-5 but the elements will be 12 but equally linear spaced
print(lsp)
print(lsp.dtype)
print(lsp.shape)
print(lsp.size)

lsp2 = np.linspace(1,50,10) #the range will be 1-50 but the elements will be 10
print(lsp2)


#creating an array with np.empty()

emp = np.empty((1,5)) #gives garbage values available in the memory, faster than np.zeros, np.ones 
print(emp)


#creating an array with np.reshape()

arr3 = np.arange(20) #gives normal range of numbers from 0 to 19
print(arr3)

newarr3 = arr3.reshape(5,4) #reshapes the array with 5 rows and columns and it gives view of an array 
print(newarr3)

#creating an array with np.full()

arr4 = np.full((2,2),7) #it takes the shape and the number that we want to fill our matrix with
print(arr4)

#creating an identity matrix using np.eye()

arr5 = np.eye(3) #gives a 3x3 matrix by default 
print(arr5)
```

## 2. Dimensions In Arrays

- A dimension in arrays is one level of array depth (nested arrays).

### Types Of Dimesional Arrays

- 0D Array: The elements in an array
- 1D Array: An array that has 0-D arrays as its elements is called uni-dimensional or 1-D array. These are the most common and basic arrays. 
- 2D Array: An array that has 1-D arrays as its elements is called a 2-D array. These are often used to represent matrix or 2nd order tensors

```python
#making a 0D array 

import numpy as np

arr3 = np.array(40)
print(arr3)
print(arr3.ndim) #ndim gives no of dimensions 
print(arr3.shape) #shape gives the size along each dimension in a tuple 


#making a 1D array

arr4 = np.array([40, 10, 30])
print(arr4)
print(arr4.ndim)
print(arr4.shape) 



#making a 2D (Matrix) array 

arr5 = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])

print(arr5)
print(arr5.ndim)
print(arr5.shape)
```

### Accessing 2D Arrays

- To access elements from 2-D arrays we can use comma separated integers representing the dimension and the index of the element.

- Think of 2-D arrays like a table with rows and columns, where the dimension represents the row and the index represents the column.

- Syntax: [rowindex, columnindex]


### Concept Of Axis In Accessing 

- Rows: Axis 0 means if axis is 0 all the operations will be done along column elements

- Columns: Axis 1 means if axis is 1 all the operations will be done along row elements

- Remember:

- 1D arrays have only axis 0 means only rows 

- 2D arrays have both axis 0 and axis 1 

```python
import numpy as np 

x = [[1,2,3], 
     [4,5,6], 
     [7,1,0]]

myarr = np.array(x)
print(myarr)

myarr.sum(axis=0) #it has summed the column elements 
myarr.sum(axis=1) #it has summed the row elements 

myarr.T #it has transposed the array

myarr.argmax() #gives max element in array
myarr.argmax(axis=0)
myarr.argmin() #gives min element in array
myarr.argsort() #gives the indices if the elements were at these indices, they would have been sorted 
myarr.argmax(axis=0) #check max elements along columns
myarr.argmin(axis=1) #checks min elements along rows 
myarr.ravel() #it ravels the elements in a line 
myarr.reshape(9,1)
```

# 7. Copy Vs View In Arrays 

- Copy: It is the new array, the copy owns the data and any changes made to the copy will not affect original array, and any changes made to the original array will not affect the copy.

- View: It is the view of original array, the view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view. 

```python
#demonstrating copy()

import numpy as np

arr12 = np.array([1, 2, 3, 4, 5])

x = arr12.copy() #here we have made changes in copy() that wont change original one 
arr12[0] = 100

print(arr12)
print(x)


#demostrating view()

arr13 = np.array([1, 2, 3, 4, 5])

y = arr13.view()
arr13[1] = 200

print(arr13)
print(y) #this has changed original too, coz we made changes in view()

```

### Check If An Array Owns Its Data with .base

- Every NumPy array has the attribute base that returns None if the array owns the data.

- Otherwise, the base  attribute refers to the original object.

```python
#demonstrating copy()

import numpy as np

arr12 = np.array([1, 2, 3, 4, 5])

x = arr12.copy() #here we have made changes in copy() that wont change original one 
arr12[0] = 100

print(arr12)
print(x)


#demostrating view()

arr13 = np.array([1, 2, 3, 4, 5])

y = arr13.view()
arr13[1] = 200

print(arr13)
print(y) #this has changed original too, coz we made changes in view()

print(x.base) #x is not changed means it owns its data
print(y.base) #y is changed means it doesnot own its data 
```

# 9. Array Operations

- Joining Arrays: It can be done via stack(), hstack(), vstack()

### Joining Arrays

- Stacking is same as concatenation, the only difference is that stacking is done along a new axis.

- We can concatenate two 1-D arrays along the second axis which would result in putting them one over the other, ie. stacking.

- We pass a sequence of arrays that we want to join to the stack() method along with the axis. If axis is not explicitly passed it is taken as 0.

```python
#normal stacking using stack()

arr16 = np.array([1, 2, 3])

arr17 = np.array([4, 5, 6])

ar2 = np.concatenate([arr16, arr17]) #concatenates the arrays 
print(ar2)

newarr = np.stack((arr16, arr17), axis = 1)
print(newarr)


#stacking along rows using hstack()

arr16 = np.array([1, 2, 3])

arr17 = np.array([4, 5, 6])

newarr2 = np.hstack((arr16, arr17))
print(newarr2)


#stacking along columns using vstack()

arr16 = np.array([1, 2, 3])

arr17 = np.array([4, 5, 6])

newarr3 = np.vstack((arr16, arr17))
print(newarr3)


#stacking along height (depth) using dstack()

arr16 = np.array([1, 2, 3])

arr17 = np.array([4, 5, 6])

newarr4 = np.dstack((arr16, arr17))
print(newarr4)

```

#### Search Sorted

- There is a method called searchsorted() which performs a binary search in the array, and returns the index where the specified value would be inserted to maintain the search order.

- The searchsorted() method is assumed to be used on sorted arrays.

```python
#here we are using a sorted array

arr = np.array([6, 7, 8, 9])

newarr7 = np.searchsorted(arr, 7) #The number 7 should be inserted on index 1 to remain the sort order.

#The method starts the search from the left and returns the first index where the number 7 is no longer larger than the next value.
print(newarr7)

```

### Searching For Multiple Values

- To search for more than one value, use an array with the specified values.

```python
#it is showing the indices where 2,4,6 should be inserted to maintain order in the array

arr = np.array([1, 3, 5, 7])

x = np.searchsorted(arr, [2, 4, 6])
print(x)
```

### Inserting In Arrays

## 10. Broadcasting & Vectorization In Arrays

- Broadcasting: NumPy automatically expands smaller arrays to larger arrays so shapes match.

- Vectorization: Doing operations on whole arrays without loops at once, easy to apply on matrix ops

### 1. Rules Of Broadcasting 

- Matching Dimensions 

- Eg: [1,2,3] + [4,5,6] = [5,7,9]

- Expanding Single Elements 

- Eg: [1,2,3] + 10 = [11,12,13]

- Incompatible Shapes

- Eg: [1,2,3] + [1,2] -> gives value error coz shapes are different 

```python
#broadcasting concept

a = np.array([1,2,3])
print(a+5)


prices = np.array([100, 200, 300])
discount = 10 #this is scalar 

finalprices = prices - (prices*discount/100)
print(finalprices)
```

```python
#vectorization concept
#fast C Loop arithmetical ops

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a+b)


ar3 = np.array([10,20,30])
multiplied = ar3*5
print(multiplied)
```

## 11. Handling Missing Values

- np.isnan()

- np.nan_to_num()

- np.isinf()

## (A) Finding Missing Values

### 1. np.isnan()

- It is a NumPy function that checks whether each element in an array is NaN (Not a Number).

- Returns True for values that are NaN

- Returns False for all other numbers

```python
#true means that there are missing values present at these indices 
#we cannot compare NaN values 

ar4 = np.array([1,2,np.nan, 3, 4, np.nan, 5,6, np.nan])
print(np.isnan(ar4)) 
```

## (B) Converting Missing Values To Numbers

### 2. np.nan_to_num()

- It converts a missing value to a number that we specify, default value is 0 always

```python
#using default value 0

ar5 = np.array([1, 2, np.nan, 3, 4, np.nan, 5, 6, np.nan])

cleanar5 = np.nan_to_num(ar5)
print(cleanar5)

#using a specified value nan=100

cleanar5 = np.nan_to_num(ar5, nan=100)
print(cleanar5)
```

## (C) Finding Infinite Values


### 3. np.isinf()

- It checks whether there are infinite values in data 

- Infinite values can be 1/0, np.inf, -np.inf

```python
ar6 = np.array([1, 2, -np.inf, 3, 4, 5, 6, np.inf, 7, 8])
print(np.isinf(ar6))
```

## (D) Converting Infinite Values To Finite Ones


### 4. np.nan_to_num()

```python
ar7 = np.array([1, 2, -np.inf, 3, 4, 5, 6, np.inf, 7, 8])
print(ar7)

cleanar7 = np.nan_to_num(ar7, posinf=1000, neginf=-1000)
print(cleanar7)
```
