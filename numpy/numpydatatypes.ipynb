# 1. DataTypes In Python 

- strings - used to represent text data, the text is given under quote marks. e.g. "ABCD"
- integer - used to represent integer numbers. e.g. -1, -2, -3
- float - used to represent real numbers. e.g. 1.2, 42.42
- boolean - used to represent True or False.
- complex - used to represent complex numbers. e.g. 1.0 + 2.0j, 1.5 + 2.5j

# 2. DataTypes In NumPy 

- i - integer
- b - boolean
- u - unsigned integer
- f - float
- c - complex float
- m - timedelta
- M - datetime
- O - object
- S - string
- U - unicode string
- V - fixed chunk of memory for other type ( void )

# 3. Checking DataType Of An Array With Using dtype()

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print(arr.dtype) #here it gives int 64 means that each element is stored in 64 bits 

arr2 = np.array([2.0, 14.89, 23.67, 34.20])
print(arr2.dtype)

arr3 = np.array(['apple', 'banana', 'cherry'])
print(arr3.dtype)
```

# 4. Creating Strings With A Defined DataType 

- We use the array() function to create arrays, this function can take an optional argument: dtype that allows us to define the expected data type of the array elements:


```python
#NumPy converts all elements to byte strings.
#It checks the length (in bytes) of each element.
#The largest length among all elements becomes the fixed byte size for the entire array.
#Every element is then padded if needed to fit this size.
#we cannot add anything to the string coz its bytesize 
#to make strings, we use unicode string 

arr4 = np.array([1,2,3,4,5], dtype= "S") 
print(arr4.dtype)


arr5 = np.array([10.3, 23.98, 45.12, 76.34], dtype = "S") 
print(arr5.dtype)

arr6 = np.array([10, 20, 30, 40.56], dtype = "S")
print(arr6.dtype)
```
