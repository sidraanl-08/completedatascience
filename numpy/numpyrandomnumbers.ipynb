# 3. Generating Random Numbers In Numpy 

- NumPy has a special random module to generate random numbers 



```python
from numpy import random

#generating random integers with randint()

x = random.randint(10)
print(x) #now with every click it is generating a random integer between 0-1


#generating random float with rand()

y = random.rand()
print(y)


#generating random 1D array with randint()

z = random.randint(100, size=(5)) #here we kept range from 0-10, and no of integers is 5 
print(z)

#generating a random 2D array 

arr = random.randint(10, size=(3,5)) #here the rows are 3 with 5 integers
print(arr)


#generating 1D array with random floats 

arr2  = random.rand(5)
print(arr2)

#generating 2D array with random floats

arr3 = random.rand(3,5)
print(arr3)
```

# 4. Generating Random From An Array

- We use choice() for a given array to get a random number

- The choice() method takes an array as a parameter and randomly returns one of the values.

```python
#using choice() to get a random number 

arr4 = random.choice([23, 45, 21, 10, 67, 98])
print(arr4) #now its generating random numbers but from a given array


#using choice() to get an array of values combined with size

arr5 = random.choice([3,5,7,9], size=(3,5))
print(arr5)
```

# 5. Random Data Distribution

- Data Distribution is a list of all possible values, and how often each value occurs.

- Such lists are important when working with statistics and data science.

- The random module offer methods that returns randomly generated data distributions.

## 1. Random Distribution: 

- A random distribution is a set of random numbers that follow a certain probability density function.

### Probability Density Function

- A function that describes a continuous probability. i.e. probability of all values in an array.

### Generating Random Numbers Based On Defined Probabilites

- We can generate random numbers based on defined probabilities using the choice() method of the random module.

- The choice() method allows us to specify the probability for each value.

- The probability is set by a number between 0 and 1, where 0 means that the value will never occur and 1 means that the value will always occur.

```python
#we have used probabilites here
#it means 3 has 10% chance to appear 
#it means 5 has 30% chance
#it means 7 has 60% chance 
#it means 9 has 0% chance thats why it wont appear among the numbers 
#the sum of all probabilites should be 1, else numpy will give probability error

x = random.choice([3,5,7,9], p=[0.1, 0.3, 0.6, 0.0], size=(100))
print(x)


#generating a 2D array same way 
#now its generating 1,2,3,4 in 3 rows with 5 numbers, but according to probability chances

y = random.choice([1,2,3,4], p=[0.1, 0.2, 0.3, 0.4], size=(3,5))
print(y)
```

# 6. Random Permutations

- A permutation refers to an arrangement of elements. e.g. [3, 2, 1] is a permutation of [1, 2, 3] and vice-versa.

- The NumPy Random module provides two methods for this: shuffle() and permutation().

### 1. Shuffling Arrays

- Shuffle means changing arrangement of elements in-place. i.e. in the array itself.

```python
#shuffle makes changes in the original array

from numpy import random
import numpy as np

arr = np.array([1,2,3,4,5])
random.shuffle(arr)
print(arr)
```

### 2. Permutating Arrays

- The permutation() method returns a re-arranged array (and leaves the original array un-changed).
 

```python
#permutation gives another array and leaves the original array unchanged

arr2 = np.array([2,4,6,8,10])
print(random.permutation(arr2))
print(arr2)
```

## 1. Normal Distribution (Gaussian Distribution)

- It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gauss.

- It fits the probability distribution of many events, eg. IQ Scores, Heartbeat et

### Parameters:

- loc - (Mean) where the peak of the bell exists.

- scale - (Standard Deviation) how flat the graph distribution should be.

- size - The shape of the returned array.

```python
#making normal distribution with size only 

from numpy import random

x = random.normal(size=(2,3))
print(x)


#making normal distribution with loc, scale, size

y = random.normal(loc=1, scale=2, size=(2, 3))
print(y)


#visualizing normal distribution 
#we get a bell shape curve for normal distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size=1000), kind="kde")
plt.show()
```

## 2. NumPy Binomial Distribution

- Binomial Distribution is a Discrete Distribution.

### Discrete Distribution:

- Discrete Distribution:The distribution is defined at separate set of events, e.g. a coin toss's result is discrete as it can be only head or tails whereas height of people is continuous as it can be 170, 170.1, 170.11 and so on.

- It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails.

### Parameters:

- n - number of trials.

- p - probability of occurrence of each trial (e.g. for toss of a coin 0.5 each).

- size - The shape of the returned array


```python
#given 10 trials of for a coin toss, generate 10 data points

from numpy import random

x = random.binomial(n=10, p=0.5, size=10)
print(x)


#visualizing binomial distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n=10, p=0.5, size=1000))

plt.show()
```

### Difference Between Normal & Binomial Distribution

- The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

## 3. Poisson Distribution

- Poisson Distribution is a discrete distribution

- It estimates how many times an event can happen in a specified time. e.g. If someone eats twice a day what is the probability he will eat thrice?

### Parameters: 

- lam - rate or known number of occurrences e.g. 2 for above problem.

- size - The shape of the returned array.

```python
#generating a random 1x10 distribution for occurrence 2

from numpy import random

x = random.poisson(lam=2, size=10)

print(x)


#visualizing poisson distribution 


from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.poisson(lam=2, size=1000))

plt.show()
```

### Difference Between Normal & Poisson Distribution

- Normal distribution is continuous whereas poisson is discrete.

- But we can see that similar to binomial for a large enough poisson distribution it will become similar to normal distribution with certain std dev and mean.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(loc=50, scale=7, size=1000),
  "poisson": random.poisson(lam=50, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

### Difference Between Binomial & Poisson Distribution

- Binomial distribution only has two possible outcomes, whereas poisson distribution can have unlimited possible outcomes.

But for very large n and near-zero p binomial distribution is near identical to poisson distribution such that n * p is nearly equal to lam.


```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "binomial": random.binomial(n=1000, p=0.01, size=1000),
  "poisson": random.poisson(lam=10, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```
