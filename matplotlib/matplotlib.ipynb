## MatPlotlib Terms We Should Know:

1. Data Point: simply referring to x and y axis
2. x-axis, y-axis: the horizontal and vertical lines on the graph used for plotting
3. Figure: canvas on which different graphs are made
4. Axes: Graph boxes, or an rectangular area where the graph is ploted
5. Plot: The lines, bars, shapes that are drawn in the axes
6. Marker: Symbol or dot used to make datapoints visible
7. Line Style: Customizing lines on the graph 
8. Legend: It is a small box that shows which line and color represents what in the graph
9. Label: The names of x-axis and y-axis on the graph like sales/year
10. Grid: Horizontal and vertical lines added behind the graphs
11. Function: Block of codes performing a speicifc task
12. Method: Functions that belong to an object
13. Parameters: Information passed to the functions
14. Keyword Arguments Kwargs: 
15. DPI Dots Per Inch: Resolution of graph controlled by dpi
16. Backend: TKagg is default, Qt5agg, Webagg for web apps

## Understanding Basic Functions of pyplot()


# PyPlot API

- Univariate: When we focus on one variable only like in numerical column
- Univariate: When we focus on one variable only like in categorical column


```python
#demostrating univariate: Numerical, histograms, boxplots

data = {
    "salary":[1000, 3300, 2200, 5400]
}

df = pd.DataFrame(data)
print(df)
print(df.head())


#demonstrating line plot

plt.plot(df['salary'], color="red", marker="o", linestyle=":", linewidth="2" )
plt.grid()
plt.show()


#demonstrating histogram

plt.hist(df['salary'], bins=5, color="green", edgecolor="black")
plt.show()


#demonstrating boxplot (good for identifying outliers)
#boxplot without outliers

plt.boxplot(df['salary'])
plt.show()


#boxplot with outliers
#df.loc[10] = [-10000]
#plt.boxplot(df['salary'])
#plt.show()


#demonstrating univariate: Categorical

df['dept'] = ['HR', 'IT', 'Finance', 'Sales', ]
print(df)
```

```python
import matplotlib.pyplot as plt

x = ["mon", "tues", "wed", "thurs", "fri"]
y = [10, 15, 7, 20, 12]

plt.plot(x,y)

plt.title("Bakery Sales This Week")

plt.xlabel("Days of week")
plt.ylabel("Sales per day")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales= [1000, 1500, 1200, 1800]
plt.plot(months, sales, color="green", linestyle="--", linewidth = 2, marker = "o", markerfacecolor="yellow", markersize=10 ,label="2025 sales data" )
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("Monthly Sales Data")
plt.legend() #gives a descriptive block 
plt.grid(color="purple") #displays a grid behind the graph 
plt.xticks([1,2,3,4],["M1", "M2", "M3", "M4"]) #divides x axis into parts 
plt.show()
```

## Plotting Bar Charts, Pie Charts, and Histograms With MatPlotlib

- Bar Charts: For comparing values and category analysis

- Pie Charts: For proportion illustration

- Histograms: For numerical data representation speically for dataspread 




## Creating Bar Graphs (Verical & Horizontal):

```python
#demostrating vertical bar graphs

import matplotlib.pyplot as plt

product = ["A", "B", "C", "D", "E"]
sales = [1000, 1500, 800, 2000, 500]

plt.bar(product, sales, color="pink", label="sales")
plt.xlabel("products")
plt.ylabel("sales")
plt.title("product sale comparison")
plt.legend()
plt.show()


#demonstrating vertical bar graphs

product = ["A", "B", "C", "D", "E"]
sales = [1000, 1500, 800, 2000, 500]

plt.barh(product, sales, color="blue", label="sales")
plt.xlabel("products")
plt.ylabel("sales")
plt.title("product sale comparison")
plt.legend()
plt.show()
```

## Creating Pie Chart:

```python
import matplotlib.pyplot as plt

regions = ["North", "South", "East", "West"]
revenue  = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct = "%1.1f%%", colors=["green", "yellow", "red", "purple"])
plt.title("Revenue Contribution By Region")
plt.show()
```

## Creating Histograms:

```python
import matplotlib.pyplot as plt

scores = [45, 53, 23, 15, 67, 34, 89, 82, 73, 65, 92, 80, 96, ]

plt.hist(scores, bins=5, color="magenta", edgecolor="black" )
plt.xlabel("score range")
plt.ylabel("no. of students")
plt.title("Score Distribution Of Students")
plt.show()
```

## Creating Scatter Plots:

- A graph where graphs are used to find relation between 2 data points

```python
#handling one group in scatter plot

import matplotlib.pyplot as plt

hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50, 55, 63, 70, 80, 85, 90, 95]

plt.scatter(hours_studied, exam_scores, color="red", marker="^", label="student data")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Scores")
plt.title("Hours Studied vs Exam Scores")
plt.legend()
plt.grid(True)
plt.show()


#handling two groups in scatter plot

plt.scatter([1,2,3], [50, 55, 60], color="Orange", label="Class A")
plt.scatter([1,2,3], [45, 50, 52], color="Blue", label="Class B")

plt.xlabel("Hours Studied")
plt.ylabel("Exam Scores")
plt.title("Comparison Of Two Classes")
plt.legend()
plt.grid(True)
plt.show()
```

## Creating SubPlots:

- Creating Parts Of Data At The Same Time 

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [ 10, 20, 15, 25 ]

plt.subplot(1,2,1)
plt.plot(x,y, color="red")
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [ 10, 20, 15, 25 ]

fig, ax = plt.subplots(1,2, figsize= (10,5))
ax[0].plot(x,y, color="green")
ax[0].set_title("Line Chart")

ax[1].bar(x,y, color="orange")
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison Of Line & Bar Charts")

plt.tight_layout()
plt.show()
```

## Saving MatPlotlib Charts:


```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10, 20, 15, 25]

plt.plot(x,y, color="red", marker="o")
plt.title("simple line plot")
plt.xlabel("X Axis")
plt.xlabel("Y Axis")

plt.savefig('lineplot.png', dpi=300, bbox_inches="tight")
plt.show()
```
