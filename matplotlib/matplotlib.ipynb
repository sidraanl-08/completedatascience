# MatPlotlib For Data Visualization


## What Is Data Visualization?

- We can visualize data and its behavior and its changes using matplotlib 

```python
import matplotlib.pyplot as plt


x = [1,2,3,4]
y = [10, 20, 15, 25]

plt.plot(x,y)
```
